# Support Vector Machine (SVM)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
cancer.keys()
```

```python
print(cancer.DESCR[:900])
```

```python
df_features = pd.DataFrame(data= cancer.data, columns= cancer.feature_names)
df_targets = pd.DataFrame(data= cancer.target, columns=['Class'])
df = pd.concat([df_features, df_targets], axis = 1)
```

```python
df.head()
```

```python
df.shape
```

```python
np.bincount(cancer.target)
```

## Splitting Data 

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_features, df_targets, test_size=0.3, random_state=101, stratify=df_targets)
```

## Scaling Data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(X_train,y_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

## Model Prediction and Evaluation

```python
from sklearn.svm import SVC
```

```python
model = SVC()
model.fit(X_train_scaled, y_train)
predictions = model.predict(X_test_scaled)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(y_test, predictions))
print('\n')
print(classification_report(y_test, predictions))
```
